```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('/content/image 3.webp')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Reshape the image to a 2D array of pixels
pixel_data = img.reshape((-1, 3))
```

```python
k = 4 # Number of clusters/colors
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixel_data)

# Replace each pixel with the nearest cluster center
quantized_pixels = kmeans.cluster_centers_[kmeans.labels_]
quantized_image = quantized_pixels.reshape(img.shape).astype('uint8')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f'Quantized Image ({k} colors)')
plt.imshow(quantized_image)
plt.axis('off')

plt.tight_layout()
plt.show()
```
